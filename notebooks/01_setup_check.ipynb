```python
import sys
import pandas as pd
import geopandas as gpd
import folium
import matplotlib.pyplot as plt
import mapclassify

print("Python:", sys.version.split()[0])
print("pandas:", pd.__version__)
print("geopandas:", gpd.__version__)
print("folium:", folium.__version__)
print("mapclassify:", mapclassify.__version__)
```

```python
from shapely.geometry import Point

sample = gpd.GeoDataFrame(
    {"name": ["Dortmund", "Bochum"]},
    geometry=[Point(7.4653, 51.5136), Point(7.2162, 51.4818)],
    crs="EPSG:4326",
)

print(sample)

sample.plot(marker="o", figsize=(4, 4))
plt.title("Sanity check: Dortmund and Bochum")
plt.show()
```

```python
m = folium.Map(location=[51.5136, 7.4653], zoom_start=10)

folium.Marker(
    [51.5136, 7.4653],
    popup="Dortmund"
).add_to(m)

m
```
